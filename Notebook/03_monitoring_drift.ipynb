# Notebook 03 – Monitoring & Data Drift avec Evidently

Hypothèse : `application_train` = données d'entraînement, `application_test` = données de production.
Ce notebook génère un rapport HTML de drift à intégrer dans le dossier de soutenance.

```python
from pathlib import Path
import pandas as pd
from evidently.report import Report
from evidently.metrics import DataDriftTable

DATA_DIR = Path('..') / 'Src' / 'Data'
OUTPUT_DIR = Path('..') / 'mlops'
OUTPUT_DIR.mkdir(exist_ok=True, parents=True)

train = pd.read_csv(DATA_DIR / 'application_train.csv')
test = pd.read_csv(DATA_DIR / 'application_test.csv')

report = Report(metrics=[DataDriftTable()])
report.run(reference_data=train, current_data=test)
report.save_html(OUTPUT_DIR / 'data_drift_report.html')
print('Report saved to', OUTPUT_DIR / 'data_drift_report.html')
```
